```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.csv')
```

```python
print("Shape:", df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
print("\nPercentage missing:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Age — mean se bharo
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Embarked — mode se bharo
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Cabin — drop karo
df = df.drop(columns=['Cabin'])

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

```python
# Sex column ko number mein badlo (male=1, female=0)
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})

# Embarked ko number mein badlo
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

print("Data types after conversion:")
print(df.dtypes)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Age aur Fare ko 0-1 ke beech scale karo
df['Age'] = scaler.fit_transform(df[['Age']])
df['Fare'] = scaler.fit_transform(df[['Fare']])

print("Age and Fare after scaling:")
print(df[['Age', 'Fare']].head())
```

```python
# IQR method se outliers hatao
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Outlier boundary
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Outlier range: {lower:.2f} to {upper:.2f}")
print(f"Outliers found: {((df['Fare'] < lower) | (df['Fare'] > upper)).sum()}")

# Outliers clip karo (remove karne ki jagah limit pe rakh do)
df['Fare'] = df['Fare'].clip(lower, upper)
```

```python
# Clean CSV save karo
df.to_csv('titanic_clean.csv', index=False)
print("Clean dataset saved!")
print("Final shape:", df.shape)
print(df.head())
```
